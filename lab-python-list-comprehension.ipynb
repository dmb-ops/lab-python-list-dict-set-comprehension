````python
# Function to initialize inventory using dictionary comprehension
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

# Function to get customer orders using set comprehension
def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {input("Enter the name of a product that a customer wants to order: ").strip().lower()
                       for _ in range(num_orders)}
    return customer_orders

# Function to update inventory and remove zero-quantity products
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
    # Use dictionary comprehension to filter out zero-quantity products
    inventory = {product: qty for product, qty in inventory.items() if qty > 0}
    return inventory

# Function to calculate order statistics
def calculate_order_statistics(customer_orders, products):
    total_ordered = len(customer_orders)
    total_products = len(products)
    percentage_unique = (total_ordered / total_products) * 100 if total_products > 0 else 0
    return total_ordered, percentage_unique

# Function to print order statistics
def print_order_statistics(order_statistics):
    total_ordered, percentage_unique = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_unique:.1f}")

# Function to print updated inventory using comprehension-style formatting
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    [print(f"{product}: {qty}") for product, qty in inventory.items()]

# Function to calculate total price using comprehension
def calculate_total_price(customer_orders):
    total_price = sum(float(input(f"Enter the price of {product}: ")) for product in customer_orders)
    return total_price

# MAIN PROGRAM
def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    inventory = update_inventory(customer_orders, inventory)

    order_stats = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_stats)
    print_updated_inventory(inventory)

    total_price = calculate_total_price(customer_orders)
    print(f"\nTotal Price: {total_price:.2f}")

# Run the program
main()
````
